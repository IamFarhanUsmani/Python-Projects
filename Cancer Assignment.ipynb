## pctsomecol18_24 which is Percent of county residents ages 18-24 highest education attained: some college has 75% of null value which is too much of null value.
## Thus we will drop it, to get a clear data.

```python
df_cancer = df_cancer.drop(['pctsomecol18_24'], axis = 1)
```

```python
plt.figure(figsize=(17,7))
sns.heatmap(df_cancer.isnull(),)
plt.show()
```

## pctemployed16_over has a 4% null value, thus we will replace the null values with median to achive a workable data

```python
median_pct = df_cancer['pctemployed16_over'].median()
df_cancer['pctemployed16_over'].replace(np.nan,median_pct,inplace =True)
```

```python
plt.figure(figsize=(17,7))
sns.heatmap(df_cancer.isnull(),)
plt.show()
```
